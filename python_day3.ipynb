```python
import numpy as np
#creating array
arr1 = np.array([1, 2, 3])
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print(arr1 + arr2)

#array operation(vectorization)
print(arr1*2)
print(arr1+arr2)

#reshaping
arr3=np.arange(12).reshape(3,4)
print(arr3)

#matrix operation
matrix1=np.array([[1,2],[3,4]])
matrix2=np.array([[5,6],[7,8]])
print(np.dot(matrix1,matrix2))

#statistical operation
data=np.random.normal(0,1,1000)#1000random number from normal distribution
print(f"Mean: {data.mean()},std :{data.std()}")
```

```python
import numpy as np
import pandas as pd
data={
    'name':['alice','bob','charlie','david'],
    'age':[25,35,40],
    'salary':[5000,45000,60000,70000]
}
df=pd.DataFrame(data)
print(df)

print(df.head(2))
print(df.describe())
print(df['age'].mean())

print(df[df['age']>30])
print(df.loc[1:2, ['name','salary']])

#handling missing data
df_with_nan=pd.DataFrame({
    'a': [1,2,np.nan,4],
    'b': [5,np.nan,np.nan,8]
})
print(df_with_nan.fillna(value=0))#replace nan with 0
print(df_with_nan.dropna())#drop row with nan

df['department']=['hr','tech','tech','hr']
print(df.groupby('department').mean())
```

```python
import pandas as pd
import numpy as np

data={
    'name':['alice','bob','charlie','david'],
    'age':[25,90,35,40],
    'salary':[50000,60000,70000,80000]
}
df=pd.DataFrame(data)
print(df)

print(df.head(2))
print(df.describe())
print(df['age'].mean())

print(df[df['age']>30])
print(df.loc[1:2,['name','salary']])

df_with_nan=pd.DataFrame({
    'a':[1,2,np.nan,4],
    'b':[5,np.nan,np.nan,8]
})
print(df_with_nan.fillna(value=0))
print(df_with_nan.dropna())

df['department']=['hr','tech','tech','hr']
print(df.groupby('department').mean(numeric_only=True))

df2=pd.DataFrame({
    'name':['alice','bob','eve'],
    'bonus':[2000,1500,3000]
})
print(pd.merge(df,df2,on='name',how='left'))

date_rng=pd.date_range(start='1/1/2023',end='1/10/2023',freq='D') 
ts=pd.DataFrame(date_rng,columns=['data'])
ts['value']=np.random.randint(0,100,size=(len(date_rng)))
ts.set_index('data',inplace=True)
print(ts.resample('3D').mean())
```
